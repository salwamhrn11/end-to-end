```python
import pandas as pd
import numpy as np
```

```python
dataUniversitas = pd.read_csv("pddikti.csv")
dataJob = pd.read_csv("dataJob.csv")
```

```python
dataUniversitas
```

```python
dataUniversitas = dataUniversitas[~dataUniversitas['tahun'].isin([2020, 2021])]
```

```python
dataUniversitas
```

```python
dataJob
```

```python
#replace jd jawa barat
dataJob['location'].replace(['Bandung, JB, Indonesia','Bekasi, JB, Indonesia','Depok, JB, Indonesia','JB, Indonesia','JB, Indonesia','Karawang, JB, Indonesia'], 'Jawa Barat', inplace=True)

#replace jd jawa tengah
dataJob['location'].replace(['Salatiga, JT, Indonesia','Surakarta, JT, Indonesia'], 'Jawa Tengah', inplace=True)

#replace jadi di yogyakarta
dataJob['location'].replace(['Bantul, YO, Indonesia','Yogyakarta, YO, Indonesia','Yogyakarta, Indonesia'], 'DI Yogyakarta', inplace=True)

#replace jadi jawa timur
dataJob['location'].replace(['Gresik, JI, Indonesia','JI, Indonesia','Malang, JI, Indonesia','Surabaya, JI, Indonesia'], 'Jawa Timur', inplace=True)

#replace jadi bali
dataJob['location'].replace(['Bali, Indonesia','Badung, BA, Indonesia','BA, Indonesia','Denpasar, BA, Indonesia','Kuta, BA, Indonesia'], 'Bali', inplace=True)

#replace jadi kalimantan timur
dataJob['location'].replace(['Balikpapan, KI, Indonesia'], 'Kalimantan Timur', inplace=True)

#replace jadi kalimantan selatan
dataJob['location'].replace(['Banjarmasin, KS, Indonesia','KS, Indonesia'], 'Kalimantan Selatan', inplace=True)

#replace jadi kepulauan riau
dataJob['location'].replace(['Batam, KR, Indonesia'], 'Kepulauan Riau', inplace=True)

#replace jadi dki jakarta
dataJob['location'].replace(['Jakarta, Indonesia','Jakarta, JK, Indonesia'], 'DKI Jakarta', inplace=True)

#replace jd lampung
dataJob['location'].replace(['LA, Indonesia'], 'Lampung', inplace=True)

#replace jd Banten
dataJob['location'].replace(['Tangerang, BT, Indonesia'], 'Banten', inplace=True)
```

```python
fullData = pd.merge(dataJob, dataUniversitas, left_on='location', right_on='Provinsi', how='outer')
fullData.drop(columns=['Provinsi'], axis=1, inplace=True)
```

```python
fullData
```

```python
fullData = fullData.dropna(subset=['location'])
fullData.drop(['benefits', 'emails', 'is_remote', 'currency', 'num_urgent_words', 'job_url', 'web-scraper-start-url'], axis=1, inplace=True)
```

```python
fullData = fullData[fullData['interval'] != 'interval']
fullData.reset_index(drop=True, inplace=True)
```
